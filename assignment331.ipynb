```python
#READ THE CSV FILE 
import pandas as pd
import numpy as np

df = pd.read_csv("https://cse331.netlify.app/assignment1/customer_data.csv")
# VIEW THE FIRST FIVE ROWS OF THE DATA
df.head(5)
```

```python
#number of rows and columns
df.shape
```

```python
#maximum min and mean of age of customers
df["age"].agg(["mean","min","max"])
```

```python
#most common three names
df["name"]=df["prefix"].astype(str)+" "+df["first"].astype(str)+" "+df["last"].astype(str)
df["name"].value_counts(sort="desc").head(3)
```

```python
#customers with same phone number
df.loc[df['phone'].duplicated(keep=False),['phone','name']]


#df[df["phone"].duplicated()==True]
```

```python
#number of structural engineers
(df["profession"]=="Structural Engineer").sum()
```

```python
#number of male sturture engineers
((df["gender"]=="Male")&(df["profession"]=="Structural Engineer")).sum()
```

```python
#female structural engineers lives in alberta
df[(df["gender"]=="Female")&(df["profession"]=="Structural Engineer")&(df["province"]=="AB")]
```

```python
#average spending amount min max mean 
df["price(CAD)"].agg(["min","max","mean"])
```

```python
#Who did not spend anything
df[df["price(CAD)"]==0]
```

```python
#who spent 100 cad or more 
df.loc[df["price(CAD)"]>=100,["name"]]
```

```python
#email with credit card number
df.loc[df['cc_no'] == 5020000000000230,['email']]
```

```python
#expiring in 2019
df['cc_exp'] = pd.to_datetime(df['cc_exp'], format='%m/%Y', errors='coerce')
df['cc_exp'].dt.year
[(df['cc_exp'] < '2019-01').sum()]
```

```python
#use visa card
(df['cc_type'] == 'Visa').sum()
```

```python
#customers spent 100 cad using visa
df.loc[(df["cc_type"]=="Visa")&(df["price(CAD)"]==100),"name"]
```

```python
#two most common profession
df["profession"].value_counts(sort="desc").head(2)
```

```python
#most famous emailproviders
df['email_domain'] = df['email'].str.split('@').str[1]
df['email_domain'].value_counts().head(5)
```

```python
#emails with am.edu
df.loc[(df['email_domain'] == 'am.edu'),['name']]
```

```python
#which day have most customers
df['weekday'].value_counts().head(1)
```
